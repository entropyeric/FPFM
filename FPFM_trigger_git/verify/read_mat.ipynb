```python
import scipy.io
import numpy as np
import matplotlib.pyplot as plt
import os
import pandas as pd
import datetime
from scipy import interpolate

# 读取.mat文件
file_path = r'C:\Users\wenzh\jupyter\paradigm\FPFM_trigger\mat_data\FinForR_20251023-1.mat'
mat_data = scipy.io.loadmat(file_path)

# 提取所需数据
timestamps = mat_data['timestamps'].flatten()
pressures = mat_data['sensor_data'].flatten()
trigger_stamp = mat_data['trigger_data'].flatten()

# 将Unix时间戳转换为datetime对象
base_time = datetime.datetime.fromtimestamp(timestamps[0])
timestamps_dt = [base_time + datetime.timedelta(seconds=float(ts - timestamps[0])) for ts in timestamps]

# 设置全局Y轴限制
Y_LIMIT = (0, 400)

# 定义事件类型映射（修改为更有意义的名称）
event_types = {4: "Full", 3: "80%", 2: "40%", 1: "20%"}

# 统一设置折线图尺寸和字体大小
FIG_SIZE = (8, 6)
FONT_SIZE = 10
```

```python
plt.figure(figsize=FIG_SIZE)
plt.rcParams.update({'font.size': FONT_SIZE})
plt.plot(timestamps_dt, pressures, label='Pressure', color='blue')
```
